```python
# 필요한 라이브러리 불러오기 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import shap
from sklearn.metrics import classification_report, confusion_matrix, roc_curve 
from imblearn.under_sampling import *
from imblearn.over_sampling import *
```

```python
# 데이터 불러오기 
defect = pd.read_csv("./defect.csv")
defect.head()
```

```python
# 기초통계량 확인하기 
defect.describe()
```

```python
# 컬럼명 확인
defect.columns
```

```python
# 불량 데이터 확률 확인하기
print('Defect',round(defect['Y'].value_counts()[1]/len(defect) * 100,2), '% of the dataset')
```

```python
# 타겟변수 시각화 
colors = ["#0101DF", "#DF0101"]
sns.countplot(x = 'Y', data = defect, palette = colors)
```

타겟변수에 대한 countplot과 defect 데이터의 비율을 따져봤을 때, 불균형한 데이터임을 알 수 있다.

```python
# KDE plot을 통해 각 변수의 분포 시각화
plt.figure(figsize=(15, 12))
for i, column in enumerate(defect.iloc[:, :-1]):  # 마지막 열 Y는 제외하고 X 변수만 사용
    plt.subplot(5, 4, i+1)
    sns.kdeplot(defect[column], shade=True)
    plt.title(f'Distribution of {column}')
plt.tight_layout()
plt.show()
```

```python
# 독립변수 간 상관관계 알아보기 
defect_X = defect.drop('Y', axis = 1)
defect_y = defect['Y']
defect_X.corr()
```

```python
# 독립변수 히트맵
plt.figure(figsize=(10,10))
sns.heatmap(defect_X.corr(),annot=True,fmt=".2f")
```

X6와 X20, X8과 X18, X12와 X19 컬럼의 상관계수가 1이고, X4, X13의 상관계수 값이 Null이다. 
이에 해당 컬럼들을 알아보고자 한다.

```python
# X6와 X20, X8과 X18, X12와 X19 컬럼들이 동일한 지 알아보기
print((defect_X["X6"]-defect_X["X20"]).unique())
print((defect_X["X8"]-defect_X["X18"]).unique())
print((defect_X["X12"]-defect_X["X19"]).unique())
```

위의 코드에서 컬럼과 컬럼을 뺀 값을 확인할 때 0인 것을 알 수 있다.
동일 컬럼 중 숫자가 큰 컬럼인 X20, X18, X19를 드롭하겠다. 

```python
# X4, X13의 값 알아보기
print(defect_X['X4'].unique())
print(defect_X['X13'].unique())
```

x4, x13은 단일값을 가진 것으로 확인되어 드롭하기로 결정했다. 
즉, **x4, x13, x18, x19, x20** 을 드롭하고 새로운 변수 defect_X_dropped를 생성하겠다. 

```python
# 컬럼 드롭 및 남은 컬럼들로 새로운 변수 만들기 
defect_X_dropped = defect_X.drop(['X20', 'X18', 'X19', 'X4', 'X13'], axis= 1)
```

```python
# 다중공선성 확인해보기 
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(
    defect_X_dropped.values, i) for i in range(defect_X_dropped.shape[1])]
vif["features"] = defect_X_dropped.columns
vif
```

```python
### 트레인, 테스트 나누기
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(defect_X_dropped, defect_y, test_size=0.3, shuffle=True, stratify=defect_y, random_state=4)
```

```python
from imblearn.combine import SMOTETomek

# SMOTETomek 적용
smote_tomek = SMOTETomek(random_state=4)
X_resampled, y_resampled = smote_tomek.fit_resample(X_train, y_train)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix

lr_model = LogisticRegression(random_state = 4)
lr_model.fit(X_resampled, y_resampled)

# 예측 및 평가
y_pred_lr = lr_model.predict(X_test)

#f1_score
from sklearn.metrics import f1_score
f1 = f1_score(y_test, y_pred_lr)
print(f"f1_score: {round(f1,4)}")

# G-Mean
from imblearn.metrics import geometric_mean_score
g_mean = geometric_mean_score(y_test, y_pred_lr) # 예측값, 실제값
print(f"G-Mean: {round(g_mean,4)}")

# 혼동 행렬 시각화
cm = confusion_matrix(y_test, y_pred_lr)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier

# 모델 훈련
rf_model = RandomForestClassifier(random_state=4)
rf_model.fit(X_resampled, y_resampled)

# 예측 및 평가
y_pred_rf = rf_model.predict(X_test)

#f1_score
from sklearn.metrics import f1_score
f1 = f1_score(y_test, y_pred_rf)
print(f"f1_score: {round(f1,4)}")

# G-Mean
from imblearn.metrics import geometric_mean_score
g_mean = geometric_mean_score(y_test, y_pred_rf) # 예측값, 실제값
print(f"G-Mean: {round(g_mean,4)}")

# 혼동 행렬 시각화
cm = confusion_matrix(y_test, y_pred_rf)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
# XGBoost 분류 모델 생성
xgb_model = XGBClassifier(random_state=4)
xgb_model.fit(X_resampled, y_resampled)

# 테스트 데이터에 대한 예측
y_pred_xgb = xgb_model.predict(X_test)

#f1_score
from sklearn.metrics import f1_score
f1 = f1_score(y_test, y_pred_xgb)
print(f"f1_score: {round(f1,4)}")

# G-Mean
from imblearn.metrics import geometric_mean_score
g_mean = geometric_mean_score(y_test, y_pred_xgb) # 예측값, 실제값
print(f"G-Mean: {round(g_mean,4)}")

# 혼동 행렬 시각화
cm = confusion_matrix(y_test, y_pred_xgb)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()
```

```python
# 특성 중요도 시각화
feature_importance = xgb_model.feature_importances_
sorted_idx = np.argsort(feature_importance)
pos = np.arange(sorted_idx.shape[0]) + .5

plt.figure(figsize=(12, 6))
plt.barh(pos, feature_importance[sorted_idx], align='center')
plt.yticks(pos, X_train.columns[sorted_idx])
plt.xlabel('Feature Importance')
plt.title('Feature Importance in Random Forest Model')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.inspection import PartialDependenceDisplay
import shap

# 기존의 Random Forest 모델을 사용한다고 가정합니다 (rf_model)

# 1. Partial Dependence Plot (PDP)
features = [0, 1]  # 첫 번째와 두 번째 특성에 대해 PDP를 그립니다. 필요에 따라 조정하세요.
fig, ax = plt.subplots(figsize=(10, 6))
PartialDependenceDisplay.from_estimator(rf_model, X_train, features, ax=ax)
plt.title("Partial Dependence Plot")
plt.tight_layout()
plt.show()

# 2. SHAP Values
# TreeExplainer를 사용하여 SHAP 값을 계산합니다
explainer = shap.TreeExplainer(rf_model)
shap_values = explainer.shap_values(X_test)

# SHAP Summary Plot
shap.summary_plot(shap_values, X_test, plot_type="bar", plot_size=(10, 6))
plt.title("SHAP Summary Plot")
plt.tight_layout()
plt.show()

# SHAP Dependence Plot
feature_to_plot = 0  # 첫 번째 특성에 대해 그립니다. 필요에 따라 조정하세요.
shap.dependence_plot(feature_to_plot, shap_values[1], X_test, feature_names=X_test.columns)
plt.title("SHAP Dependence Plot")
plt.tight_layout()
plt.show()

# SHAP Force Plot (첫 번째 예측에 대해)
shap.force_plot(explainer.expected_value[1], shap_values[1][0,:], X_test.iloc[0,:], matplotlib=True)
plt.title("SHAP Force Plot")
plt.tight_layout()
plt.show()
```
