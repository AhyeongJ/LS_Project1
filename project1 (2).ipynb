```python
defect_X = defect.drop('Y', axis = 1)
defect_y = defect['Y']
defect_X.corr()
```

```python
#Y 포함한 모든 칼럼 상관계수 히트맵
plt.figure(figsize=(10,10))
sns.heatmap(defect_X.corr(),annot=True,fmt=".2f",cmap="Blues")
```

```python
## 
print((defect_X["X6"]-defect_X["X20"]).unique())
print((defect_X["X8"]-defect_X["X18"]).unique())
print((defect_X["X12"]-defect_X["X19"]).unique())
```

```python
print(defect_X['X4'].unique())
print(defect_X['X13'].unique())
```

**x4, x13, x18, x19, x20 드롭**

```python
defect_X_dropped = defect_X.drop(['X20', 'X18', 'X19', 'X4', 'X13'], axis= 1)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(
    defect_X_dropped.values, i) for i in range(defect_X_dropped.shape[1])]
vif["features"] = defect_X_dropped.columns
vif
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import StratifiedShuffleSplit
X_train, X_test, y_train, y_test = train_test_split(defect_X_dropped, defect_y, test_size=0.3,shuffle=True, stratify=defect_y, random_state=4)

# 결과 출력 (훈련 데이터에서 클래스 비율 확인)
print("훈련 데이터에서 클래스 분포:\n", y_train.value_counts(normalize=True))
print("테스트 데이터에서 클래스 분포:\n", y_test.value_counts(normalize=True))
```

# *  UnderSampling 
## 1. Random UnderSampling

```python
# (1) Random UnderSampling
from imblearn.under_sampling import RandomUnderSampler
from collections import Counter

under_sample = RandomUnderSampler(random_state=4)
X_under_random, y_under_random = under_sample.fit_resample(X_train,y_train)

# 결과 확인
print(f"Before Sampling: {Counter(y_train)}")
print(f"After Sampling, Random: {Counter(y_under_random)}")
```

## 2. Tomek Links

```python
# (2) Tomek Links
from imblearn.under_sampling import TomekLinks
from collections import Counter

# Tomek Links 적용
tl = TomekLinks(sampling_strategy='majority')
X_under_tomek, y_under_tomek = tl.fit_resample(X_train,y_train)

# 결과 확인
print(f"Before Sampling: {Counter(y_train)}")
print(f"After Sampling, Tomek Links: {Counter(y_under_tomek)}")
```

# * Model 
## 1. Random UnderSampling dataset
###     (1) LogisticRegression


```python
## model 1: 로지스틱, random
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
# 로지스틱 회귀 모델 생성
lr_model = LogisticRegression()
lr_model.fit(X_under_random, y_under_random)

# 테스트 데이터에 대한 예측
y_pred_lr = lr_model.predict(X_test)

# 모델 평가
accuracy = accuracy_score(y_test, y_pred_lr)
print(f"Accuracy: {accuracy:.4f}")
```

#### + LogisticRegression의 성능 지표 
혼동행렬, f1 score, Gmeans

```python
## 성능 1. 혼동행렬
print(classification_report(y_test, y_pred_lr))
cm = confusion_matrix(y_test, y_pred_lr)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()
```

```python
## 성능 2. f1 score
from sklearn.metrics import f1_score

# 데이터 스플릿으로 y_valid와 모델 예측으로 y_pred를 구한 후 실행
# y_pred만 모델에 맞게 수정
f1 = f1_score(y_test, y_pred_lr)
print(f"f1_score: {round(f1,4)}")
```

```python
## 성능 3. Gmeans
from imblearn.metrics import geometric_mean_score
g_mean = geometric_mean_score(y_test, y_pred_lr) # 예측값, 실제값
print(f"G-Mean: {round(g_mean,4)}")
```
